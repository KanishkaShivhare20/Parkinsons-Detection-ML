## Importing the Dependencies

```python
# Data & ML Libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

# Sklearn Modules
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn import svm 
from sklearn.metrics import accuracy_score

# Ignore specific warnings
import warnings
warnings.filterwarnings("ignore")
```

## Data Collection & Analysis

```python
# Load the data from CSV
df= pd.read_csv("../data/parkinsons.csv")

# Show first few rows
df.head()
```

```python
# Shape of dataset
print("Dataset shape:", df.shape)

# Column info
df.info()

# Check for missing values
print("\nMissing values:\n", df.isnull().sum())

# Statistical description
df.describe()
```

```python
# Distribution of the target class
sns.countplot(x='status', data=df, palette='viridis')
plt.title("Parkinson's Disease Status Distribution")
plt.xlabel("Status (0 = No Disease, 1 = Has Disease)")
plt.ylabel("Count")
plt.show()

# Numerical check
df['status'].value_counts()
```

1  --> Parkinson's Positive

0 --> Healthy


```python
# Compare mean feature values grouped by disease status
df.groupby('status').mean(numeric_only=True)
```

## Data Pre-Processing

### Separating the features & Target

```python
# Dropping 'name' and 'status' columns to get features
X = df.drop(columns=['name', 'status'], axis=1)

# Target column
y = df['status']
```

```python
print(y)
```

### Train - Test Split

```python
# Train-test split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2 , random_state=48)

print("X shape:", X.shape)
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

### Feature Scaling

```python
# Standardizing the features
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

## Model Training

### Support Vector Machine Model

```python
# Support Vector Machine (SVM) model with linear kernel
model = svm.SVC(kernel='linear')
model.fit(X_train, y_train)
```

## Model Evaluation

### Accuracy Score

```python
# Accuracy on training data
y_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)

# Accuracy on testing data
y_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("Training Accuracy: " + str(round(train_accuracy * 100, 2)) + "%")
print("Testing Accuracy: " + str(round(test_accuracy * 100, 2)) + "%")
```

### Building a Predictive System 

#### Input Prediction (Manually)

```python
# Example input (manual values)
input_data = (197.07600,206.89600,192.05500,0.00289,0.00001,0.00166,0.00168,0.00498,0.01098,0.09700,
              0.00563,0.00680,0.00802,0.01689,0.00339,26.77500,0.422229,0.741367,-7.348300,0.177551,
              1.743867,0.085569)

# Convert to numpy array & reshape
input_array = np.asarray(input_data).reshape(1, -1)

# Scale and predict
input_scaled = scaler.transform(input_array)
prediction = model.predict(input_scaled)

# Result
if prediction[0] == 0:
    print("✅ The Person does NOT have Parkinson's Disease")
else:
    print("⚠️ The Person HAS Parkinson's Disease")
```
